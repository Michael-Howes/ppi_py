# Auditing electronic voting

The goal is to construct a confidence interval for the proportion of people voting for each candidate using a small number of hand-labeled ballots and a large number of ballots read with an optical scanner. On Election Day in the United States, most voters use electronic or optical-scan ballots, neither of which is perfectly accurate. The data is taken from a special election in San Francisco for the Assembly District 17 seat on April 19, 2022. The candidates were David Campos and Matt Haney. This is an example of a risk-limiting audit: a statistically valid way to check the results of an election by inspecting subsets of ballots (see, e.g., (1)).

1. M. Lindeman and P. B. Stark, A gentle introduction to risk-limiting audits. IEEE Security & Privacy 10(5), 42–49 (2012).

### Import the ballots data set

Load the data. The data set contains a small number of hand-labeled votes (```Y```), corresponding predicted votes based on computer vision (```Yhat```), and predicted votes for the remaining ballots (```Yhat_unlabeled```).

````python
dataset_folder = "./data/"
data = load_dataset(dataset_folder, "ballots")
Y = data["Y"]
Yhat = data["Yhat"]
Yhat_unlabeled = data["Yhat_unlabeled"]
````

### Problem setup

Specify the error level (```alpha```), range of values for the labeled data set size (```ns```), and number of trials (```num_trials```).

Specify the ground-truth value of the estimand (reported election outcome).

````python
alpha = 0.05
n_max = Y.shape[0]  # Total number of labeled ballots
ns = np.linspace(100, n_max, 20).astype(
    int
)  # Test for different numbers of labeled ballots
num_trials = 500

# Reported result (official outcome)
reported_estimate = 0.6237
````

### Construct intervals

Form confidence intervals for all methods and problem parameters. A dataframe with the following columns is formed:
1. ```method``` (one of ```PPI```, ```Classical```, and ```Imputation```)
2. ```n``` (labeled data set size, takes values in ```ns```)
3. ```lower``` (lower endpoint of the confidence interval)
4. ```upper``` (upper endpoint of the confidence interval)
5. ```trial``` (index of trial, goes from ```0``` to ```num_trials-1```)

````python
# Run prediction-powered inference and classical inference for many values of n
results = []
for i in tqdm(range(ns.shape[0])):
    for j in range(num_trials):
        # Prediction-Powered Inference
        n = ns[i]
        rand_idx = np.random.permutation(n_max)
        _Yhat = Yhat[rand_idx[:n]]
        _Y = Y[rand_idx[:n]]

        ppi_ci = ppi_mean_ci(_Y, _Yhat, Yhat_unlabeled, alpha=alpha)

        # Classical interval
        classical_ci = classical_mean_ci(_Y, alpha=alpha)

        # Append results
        results += [
            pd.DataFrame(
                [
                    {
                        "method": "PPI",
                        "n": n,
                        "lower": ppi_ci[0],
                        "upper": ppi_ci[1],
                        "trial": j,
                    }
                ]
            )
        ]
        results += [
            pd.DataFrame(
                [
                    {
                        "method": "Classical",
                        "n": n,
                        "lower": classical_ci[0],
                        "upper": classical_ci[1],
                        "trial": j,
                    }
                ]
            )
        ]

# Imputed CI
imputed_ci = classical_mean_ci(Yhat_unlabeled, alpha=alpha)
results += [
    pd.DataFrame(
        [
            {
                "method": "Imputation",
                "n": np.nan,
                "lower": imputed_ci[0],
                "upper": imputed_ci[1],
                "trial": 0,
            }
        ]
    )
]

df = pd.concat(results, axis=0, ignore_index=True)
df["width"] = df["upper"] - df["lower"]
````

### Power experiment

For PPI and the classical approach, find the smallest value of ```n``` such that the method has power 80% against the null $H_0: \text{fraction for Matt Haney} \leq 50\%$.

````python
# Find n such that we reject H0: fraction for Matt Haney < 0.5 with probability 80% using a test at level alpha
num_experiments = 100
list_rand_idx = [np.random.permutation(n_max) for i in range(num_experiments)]

alpha_pval = 0.05


def _to_invert_ppi(n):
    n = int(n)
    nulls_rejected = 0
    # Data setup
    for i in range(num_experiments):
        rand_idx = list_rand_idx[i]
        _Yhat = Yhat[rand_idx[:n]]
        _Y = Y[rand_idx[:n]]

        ppi_ci = ppi_mean_ci(_Y, _Yhat, Yhat_unlabeled, alpha=alpha_pval)
        if ppi_ci[0] > 0.5:
            nulls_rejected += 1
    return nulls_rejected / num_experiments - 0.8


def _to_invert_classical(n):
    n = int(n)
    nulls_rejected = 0
    # Data setup
    for i in range(num_experiments):
        rand_idx = list_rand_idx[i]
        _Y = Y[rand_idx[:n]]

        classical_ci = classical_mean_ci(_Y, alpha=alpha_pval)

        if classical_ci[0] > 0.5:
            nulls_rejected += 1
    return nulls_rejected / num_experiments - 0.8
````

````python
n_ppi = int(brentq(_to_invert_ppi, 1, 1000, xtol=1)) # This rejects for any value n. Try it!
n_classical = int(brentq(_to_invert_classical, 1, 1000, xtol=1))
print(f"The PPI test requires n={n_ppi} labeled data points to reject the null.")
print(
    f"The classical test requires n={n_classical} labeled data points to reject the null."
)
````

## Comparison with theoretical power analysis


First we will do a hypothetical power analysis where we use the full data set to estimate $\rho$. Here we see good agreement between the theoretical power analysis and the output from PPI.

````python
rho = np.corrcoef(Y, Yhat)[0,1]
sigma_sq = np.var(Y)
N_max = Yhat_unlabeled.shape[0]


def compute_SE(n, N, rho, sigma_sq):
    var_ppi = sigma_sq/n * (1-N*rho**2/(n+N))
    return np.sqrt(var_ppi)

def estimate_SE(n, N):
    rand_idx = np.random.permutation(n_max)
    _Yhat = Yhat[rand_idx[:n]]
    _Y = Y[rand_idx[:n]]

    rand_idx = np.random.permutation(N_max)
    _Yhat_unlabeled = Yhat_unlabeled[rand_idx[:N]]

    ppi_ci = ppi_mean_ci(_Y, _Yhat, _Yhat_unlabeled, alpha=0.05)
    return (ppi_ci[1]-ppi_ci[0])/(2*1.96)
````

````python
Ns = np.linspace(100, 5101, 20).astype(int)
ns = np.linspace(100, 1101, 20).astype(int)
theory_SEs = np.zeros((Ns.shape[0], ns.shape[0]))
actual_SEs = np.zeros((Ns.shape[0], ns.shape[0]))
for i in range(Ns.shape[0]):
    for j in range(ns.shape[0]):
        theory_SEs[i,j] = compute_SE(ns[j], Ns[i], rho, sigma_sq)
        actual_SEs[i,j] = estimate_SE(ns[j], Ns[i])[0]
````

````python
plt.imshow(theory_SEs)
plt.colorbar()
plt.title("Theoretical SE")
plt.xticks(np.arange(len(ns))[::5], ns[::5])
plt.xlabel("n")
plt.yticks(np.arange(len(Ns))[::5], Ns[::5])
plt.ylabel("N")
plt.gca().invert_yaxis()
plt.show()

plt.imshow(actual_SEs)
plt.colorbar()
plt.title("Estimated SE")
plt.xticks(np.arange(len(ns))[::5], ns[::5])
plt.xlabel("n")
plt.ylabel("N")
plt.yticks(np.arange(len(Ns))[::5], Ns[::5])
plt.gca().invert_yaxis()
plt.show()
````

````python
error = np.abs((theory_SEs - actual_SEs)/theory_SEs).mean()*100
print(f"Average relative error: {error:.2f}%")
````

````python
ks = [0, 2, 8]
for k in ks:
    color = next(plt.gca()._get_lines.prop_cycler)['color']
    plt.plot(ns, theory_SEs[k,:], label="Theoretical SE N={}".format(Ns[k]), color=color)
    plt.plot(ns, actual_SEs[k,:],'o', label="Estimated SE N={}".format(Ns[k]), color=color)

plt.xlabel("n")
plt.ylabel("SE")
plt.legend()
plt.show()
````

We will now do a similar power analysis but where $\rho$ is now estimated on a subset of the data. This is a more accurate description of how the power analysis would be used in practice.

````python

def estimate_power(n0, n, N):
    rand_idx = np.random.permutation(n_max)
    _Yhat = Yhat[rand_idx[:n0]]
    _Y = Y[rand_idx[:n0]]
    rho_ = np.corrcoef(_Y, _Yhat)[0,1]
    sigma_sq_ = np.var(_Y)
    return compute_SE(n, N, rho_, sigma_sq_)


n0 = 100
N = 2000
ses = np.zeros((ns.shape[0], num_trials))
se_max = np.zeros(ns.shape[0])
for i in range(ns.shape[0]):
    se_max[i] = compute_SE(ns[i], N, rho, sigma_sq)
    for j in range(num_trials):
        ses[i,j] = estimate_power(n0, ns[i], N)

plt.plot(ns, ses.mean(axis=1), label="Estimated SE (mean)")
k = 50
ns_ = np.zeros_like(ses[:,:k])
ns_ += ns[:,None]
plt.scatter(ns_, ses[:,:k], alpha=0.4, s=5, label="Estimated SE")
plt.plot(ns, se_max, '--', label="Theoretical SE")
plt.xlabel("n")
plt.ylabel("Estimated SE")
plt.legend()
plt.title("Estimated SE from an initial data set of size {}".format(n0))
plt.show()
````
